$$

$$

\frac{average\_playtime}

# 2. TRANSFORMACIONES ÚTILES

Esta parte es CLAVE para el TP.

---

# A) Transformar `owners`

Problema:
viene como rango textual.

Ejemplo:

```python
"20000-50000"
```

## ✔ Transformación recomendada:

usar punto medio.

Ejemplo:

$$
owners\_mid = \frac{20000 + 50000}{2}
$$

Justificación:
permite:

* correlaciones,
* regresiones,
* métricas económicas,
* optimización.

---

# B) Crear variable “Indie”

```python
df['is_indie'] =
df['genres'].str.contains('Indie')
```

Justificación:
permite validar:

* crecimiento indie,
* diferencias de mercado,
* hipótesis temporal.

---

# C) Crear décadas/etapas históricas

La división temporal por etapas históricas permite contextualizar los cambios estructurales de la industria.

Ejemplo:

| Etapa             | Período   |
| ----------------- | --------- |
| Online temprano   | 2000–2006 |
| HD/Casual         | 2006–2013 |
| Streaming/Esports | 2013–2019 |

Justificación:

* análisis evolutivo,
* cambios estructurales,
* comparación histórica.

---

# D) Log-transform en ventas

Las ventas tienen distribución MUY sesgada.

Conviene:

$$
\log(owners\_mid + 1)
$$

Justificación:

* evitar dominancia de outliers,
* gráficos más legibles,
* correlaciones más estables.

---

# E) Revenue estimado

$$
estimated\_revenue = owners\_mid \times price
$$

IMPORTANTÍSIMO para:

* desempeño comercial,
* optimización,
* análisis económico.

---

# F) Clasificar publishers

Separar:

* grandes estudios,
* medianos,
* indies.

Puede hacerse por:

* cantidad de juegos,
* promedio de owners,
* revenue estimado.

# 4. LIMITACIONES IMPORTANTES DEL DATASET

ESTA PARTE ES FUNDAMENTAL para parecer un análisis serio.

Impacto:

  * cambios post-2020.

No son ventas oficiales exactas.

Impacto:

* afecta revenue estimado.

No hay:

* costos de desarrollo,
* presupuesto marketing,
* ganancias netas.

Entonces:
❌ no puede medirse rentabilidad real.

---

# D) Sesgo hacia juegos populares

Muchos juegos chicos tienen:

* pocos ratings,
* datos incompletos,
* poca visibilidad.

Ejemplo:

* que un juego tenga achievements NO implica que venda más.
  Puede ser:
* juegos exitosos reciben más contenido.

# F) Tags y géneros superpuestos

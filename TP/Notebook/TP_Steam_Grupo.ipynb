## 2.A. Transformación de la Variable 'owners' (Punto Medio)

**Problema identificado:** En la columna original de Owners recita la cantidad promedio de usuarios que poseen el videojuego mediante la codificacion de rango de texto (dando el ejemplo de "20000-50000"). Siendo un string, no hay posibilades para analizar las operaciones matematicas correspodientes.

**Solución aplicada:** La aplicacion de una transformacion numerica que esta basada en el "punto medio del rango". Al extender tanto el limite superior y el inferior del texto en cada juego, calculamos el promedio numerico creando la variable llamada `owners_mid`.

La fórmula matemática aplicada para cada registro es:

$$owners\_mid = \frac{\text{Límite Inferior} + \text{Límite Superior}}{2}$$

**Justificación técnica y económica:** Esta transformación nos permite tratar la variable de forma continua para:
* Evaluar **correlaciones** (ej: relación entre el precio y el volumen de usuarios).
* Desarrollar **regresiones** y proyecciones de demanda.
* Calcular **métricas económicas** aproximadas (como la estimación de ingresos brutos por juego).

````python
#Utilizando IA para correguir la definicion, la usamos para convertir a enteros los valores
def calcular_punto_medio_owners(rango_texto):
    """
    Toma un string con formato 'MIN-MAX' (ej: '20000-50000'),
    separa los valores, los convierte a enteros y devuelve su promedio.
    """
    try:
        # Utilizando el guion como un separador de texto
        limites = str(rango_texto).split('-')

        # Cambiar el rango a un entero
        limite_inferior = int(limites[0].strip())
        limite_superior = int(limites[1].strip())

        # El calculo del punto medio
        punto_medio = (limite_inferior + limite_superior) / 2
        return punto_medio
    except Exception as e:
        # Devolver NaN a cada valor nulo o formato inesperados
        return np.nan

#Para aplicar la nueva columna numerica
df_steam['owners_mid'] = df_steam['owners'].apply(calcular_punto_medio_owners)

#Y su verificacion rapida
print("Muestra de la transformación de 'owners':")
print(df_steam[['name', 'owners', 'owners_mid']].head())
````

### 📊 Interpretación Económica e Implicancias del Negocio

En el analisis de *two-sided markets*, la cantidad de los usuarios acumulados del activo estrategico y la barrera de la entrada critica. Con el dataset que provee los rangos de trasacciones por los motivos de confidencialidad comercial, la variable `owners_mid` funciona como **proxy de la demanda cuantitativa histórica (Q)** para cada firma dentro del ecosistema de Steam.

Observando la salida de codigo, los resultados reflejan el comportamiento mencionado. Titulos pioneros de Valve como es *Counter-Strike* que registran la demanda de **15,000,000 de usuarios**, al mismo tiempo los clásicos como *Team Fortress Classic* o *Day of Defeat* se estabilizan en **7,500,000**. Disponiendo los valores continuos donde permite modelar la aproximacion de la **Función de Ingresos Totales (IT)** mediante:

$$\text{Ingreso Estimado} \approx \text{price} \times owners\_mid$$

**Nota metodológica del equipo:** El equipo pudo reconocer que el indicador representa el *ingreso devengado bruto potencial* en la lista de precios actuales. La realidad de las organizaciones digitales donde el valor no deduce la tasa de intermediación de la plataforma (la comisión del 30% de Valve), donde los efectos de la dicriminacion de los precios (oferta estacionales) o las elasticidades de los precios regional tampoco.

Sin embargo, este valor analitico de la transformacion se encuentra en la capacidad para **homogeneizar y estandarizar la escala de las firmas**. Donde pasar de rangos categóricos como `"10000000-20000000"` a métricas numéricas exactas, es uno de los pasos indispernsable para el calculo de las cuotas del mercado que se evalua la concentracion de la oferta respecto a la demanda y la clasificacion del exito comercial en las secciones posteriores.

## 2.C. Creación de Variables Temporales por Etapas Históricas

**Problema identificado:** El analisis de los lanzamientos anuales donde fragmentar el analisis, ocultanto las tendencias mecro con el objetivo de entender la evolucion de la industria. Con ello hay que agrupar en periodos de tiempo donde reflejen hitos tecnologico y comercial.

**Solución aplicada:** En primer lugar hay que asegurar de extraer los años de lazamientos de la variable `release_date`. Utilizando la funcion `pd.cut()` para la caracterizacion de los videojuegos en una de las tres etapas hitoricas:

* **Online temprano (2000-2006):** Los inicios de Steam, la transición del formato físico al digital y la consolidación de las conexiones de banda ancha.
* **HD / Casual (2006-2013):** Expansión masiva del mercado, popularización de los gráficos en alta definición y la llegada del juego casual/indie a gran escala.
* **Streaming / Esports (2013-2019):** Maduración del ecosistema, explosión de los juegos competitivos, transmisiones en vivo (Twitch) y consolidación del modelo de negocio de acceso abierto con microtransacciones.

**Justificación técnica y económica:**
* Realizar un **análisis evolutivo** agregando métricas por bloques históricos.
* Detectar **cambios estructurales** en la oferta (cantidad de lanzamientos) y en la demanda (volumen de jugadores).
* Permitir una **comparación histórica** robusta sobre cómo varían los precios promedio y los ingresos según la época en que se lanzó el juego.

````python
# Con el apoyo de la IA para la correccion de errores presentados.
# 1. Definir como datetime en el 'release_date' para extraer años.
df_steam['release_year'] = pd.to_datetime(df_steam['release_date']).dt.year

# 2. Hay que definir los grupos correspodientes de cada etapa historica.
# Los años 1999 y 2019 son los extremos que los incluyen.
limites_años = [1999, 2006, 2013, 2019]
nombres_etapas = ['Online temprano', 'HD/Casual','Streaming/Esports']

# 3. Se crea la variable con pd.cut
df_steam['historical_stage'] = pd.cut(df_steam['release_year'], bins=limites_años, labels=nombres_etapas)

# 4. Forma de verificacion que cuenta la cantidad de juegos en cada etapa
print("Distribución de videojuegos por etapa histórica:")
print(df_steam['historical_stage'].value_counts().sort_index())
````

## 2.D. Transformación Logarítmica de las Ventas Estimadas (Log-Transform)

**Problema identificado:** La distribucion de `owners_mid` presenta el sesgo positivo. La mayoria de los titulos que acumula el volumen de usuarios donde un grupo de "bestsellers" concentra millones de copias. La simetria extrema dificulta la observacion de datos y distorsiona el comportamiento a causa de los valores atipicos (*outliers*).

**Solución aplicada:** La aplicacion de la tranformacion logaritmica utilizando un logaritmo natural en la variable mas uno, mediante:


$$\log\_owners = \log(owners\_mid + 1)$$

Sumar una unidad ($+1$) es una práctica matemática estándar que previene errores de indeterminación en el caso hipotético de que existan registros con valor cero, dado que $\log(0)$ no está definido, mientras que $\log(1) = 0$.

**Justificación técnica y económica:** Esta normalización de los datos es fundamental para:
* **Evitar la dominancia de outliers:** Suaviza los valores extremos de los "hits" comerciales para que no sesguen los estadísticos descriptivos globales.
* **Lograr gráficos más legibles:** Comprime las escalas visuales, permitiendo observar la dispersión y los patrones de los juegos pequeños y medianos que antes quedaban ocultos.
* **Estabilizar las correlaciones:** Acerca la distribución de la variable a una forma normal (campana de Gauss), un supuesto crítico para realizar regresiones lineales estables y calcular coeficientes de correlación confiables.

````python
# Con el apoyo de la IA para la correccion de errores presentados.
# 1. Aplicacion de la transformacion usando la función optimizada np.log1p
# np.log1p(x) calcula exactamente el logaritmo natural de (x + 1)
df_steam['log_owners'] = np.log1p(df_steam['owners_mid'])

# 2. Verificación con el fin de comprovar las estadisticas de ambas variables
print("Comparativa de estadísticas (Original vs Transformada):")
print("-" * 60)
print(df_steam[['owners_mid', 'log_owners']].describe().round(2))
````

### 📊 Interpretación Económica e Implicancias del Negocio

Con la perspectiva del análisis estadístico de mercados, con la comparacion de las metricas descriptivas aporta la evidencia latente de una **economía de "ganadores-se-llevan-todo"** (*Winners-Take-All*), característica intrínseca de los mercados de bienes digitales.

El analisis de los datos reales, la variable original `owners_mid` muestra la asimetria extrema: con la mediana (50%) y el percentil 25% se encuentran anclados en el mínimo de **10,000 usuarios**, el valor máximo escala de forma abrupta hasta los **150,000,000 de usuarios** (un outlier masivo).  La dispersion es severa que la desviación estándar ($1.32 \times 10^6$) supera por un orden de magnitud a la media ($1.34 \times 10^5$), lo que demuestra que los estadísticos tradicionales quedan completamente distorsionados por un puñado de éxitos comerciales.

La aplicación del logaritmo natural mediante `log_owners` logra estabilizar esta variabilidad de manera exitosa, donde comprime el rango total a una escala matemáticamente manejable que va de **9.21 a 18.83**, con una media armónica de **9.98** y una desviación estándar controlada de **1.34**.

En los terminos microeconómicos, el paso metodologico se tranforma en algo crucial. La transformacion de la variable de volumen a una escala logarítmica, cambiamos la naturaleza de las futuras estimaciones. En las secciones del modelo estadistico donde la regresion lineal que vincula el precio con `log_owners` no medirá variaciones en unidades absolutas de jugadores, las cuales carecen de sentido elástico debido a la brecha entre el piso de 10,000 y el techo de 150 millones. Esto permite calcular de forma directa la **elasticidad de la demanda**. Es decir, nos dotará de la capacidad de interpretar los impactos de las estrategias de precios en **variaciones porcentuales**, permitiendo un análisis predictivo robusto sobre el comportamiento del consumidor en la plataforma.

## 2.E. Cálculo del Revenue Estimado (Ingreso Bruto Total)

**Problema identificado:** La cantidad de usuarios puede dar la nocion del alcance que tiene un juego aunque no refleja el exito financiero. Un juego gratuito, que genera un ingreso directo de cero, que puede generar millores de usuarios puede ser exitoso al igual que un juego de 60 dolares.

**Solución aplicada:** Multiplicamos el punto medio de poseedores por el precio de lista del videojuego para generar la variable cuantitativa continua `estimated_revenue`:

$$estimated\_revenue = owners\_mid \times price$$

**Justificación técnica y económica:** Esta variable es el pilar fundamental del análisis de negocio en este TP porque permite:
* Evaluar el **desempeño comercial** agregado de los diferentes actores (comparar ingresos totales de firmas Indie vs. corporaciones).
* Servir de base para modelos de **optimización**, buscando qué combinación de variables maximiza el ingreso.
* Desarrollar un **análisis económico** profundo sobre cómo se distribuyen los recursos financieros dentro de la industria de plataformas digitales.

````python
# Con el apoyo de la IA para la correccion de errores presentados.
# 1. Calculo para el Revenue Estimado multiplicando la cantidad de usuarios por el precio
df_steam['estimated_revenue'] = df_steam['owners_mid'] * df_steam['price']

# 2. Identificar el Top 5 de juegos con mayor recaudación estimada para verificar el resultado con el formateo en miles de dolares
top_revenue = df_steam[['name', 'price', 'owners_mid', 'estimated_revenue']].sort_values(by='estimated_revenue', ascending=False).head(5)

print("Top 5 videojuegos con mayor Revenue Estimado en Steam:")
print("-" * 75)
for index, row in top_revenue.iterrows():
    print(f"Juego: {row['name'][:30]:<30} | Precio: ${row['price']:<5} | Dueños: {int(row['owners_mid']):<10} | Revenue: ${row['estimated_revenue']/1e6:.2f} M")
````

### 📊 Interpretación Económica e Implicancias del Negocio

Los terminos de **economía de la empresa**,  la variable `estimated_revenue` representa la aproximación al **Ingreso Total ($IT = P \times Q$)** percibido por las firmas dentro de la plataforma. Con el codigo donde aporta los datos sobre las diferentes estrategias de fijación de precios (*pricing*) pueden equilibrar los resultados financieros finales.

Con el top 5, se puede observar un fenomeno que es liderado de forma aislada por *PLAYERUNKNOWN'S BATTLEGROUNDS (PUBG)*, con un revenue estimado que rompe la escala alcanzando los **$2,024.25 millones de dólares**. El hallazgo demuestra la elasticidad de la demanda y la estructura de precios determinan el posicionamiento financiero, independientemente del volumen neto de la comunidad de usuarios.

Por eso, los sesgos asumidos, `estimated_revenue` se consolida como la métrica de optimización más robusta del dataset para desentrañar las asimetrías competitivas, analizar qué sectores capturan la mayor porción del excedente del consumidor y estudiar las barreras financieras que enfrentan las firmas menores ante los líderes del mercado.

## 2.F. Clasificación y Estratificación de Publishers (Editores)

**Problema identificado:** La columna `publisher` contiene miles de nombres de empresas distintas, lo que dificulta un análisis comparativo directo. Para entender la estructura competitiva del mercado, necesitamos agrupar a estos editores en categorías lógicas que reflejen su peso e influencia económica en la plataforma.

**Solución aplicada:** Primero, agrupamos el dataset por `publisher` para calcular sus métricas agregadas: la cantidad total de títulos lanzados, su promedio de usuarios (`owners_mid`) y su facturación acumulada (`estimated_revenue`).

Luego, utilizando una lógica de segmentación basada en el volumen de ingresos totales acumulados (o en su defecto, combinando volumen de catálogo y ventas), clasificamos a cada publisher en uno de tres niveles estratégicos:
* **Grandes estudios (Top / Major):** Firmas con catálogos masivos o éxitos comerciales de altísimo impacto que concentran la mayor parte del revenue del mercado.
* **Medianos (AA / Mid-tier):** Empresas consolidadas con un flujo estable de lanzamientos y ventas moderadas.
* **Indies / Pequeños:** Editores emergentes o desarrolladores independientes autogestionados con estructuras de mercado de menor escala.

**Justificación técnica y económica:** Esta estratificación empresarial es fundamental para:
* Describir la **estructura de mercado** (evaluar si es un oligopolio donde pocos publishers concentran todo el ingreso).
* Analizar las **estrategias de cartera de productos** según el tamaño de la firma.
* Facilitar la **comparación histórica** del rendimiento comercial de los pequeños competidores frente a las corporaciones dominantes.

````python
# Con el apoyo de la IA para la correccion de errores presentados.
# 1. La agrupacion por publisher para calcular el perfil económico
publisher_stats = df_steam.groupby('publisher').agg(
    total_games=('name', 'count'),
    avg_owners=('owners_mid', 'mean'),
    total_revenue=('estimated_revenue', 'sum')
).reset_index()

# 2. Clasificacion usando pd.qcut y .apply()
def segmentar_publishers_por_cuantil(df):
    """Segmenta los publishers en categorías (Pequeño, Mediano, Grande)

    basándose en cuantiles del 'total_revenue'.
    """
    df = df.copy()
    labels = ["Indie/Pequeño", "Mediano", "Grande"]
    # pd.qcut with duplicates="drop" will reduce the number of bins if there are not enough unique values.
    try:
        df["categoria_publisher"] = pd.qcut(
            df["total_revenue"], q=3, labels=labels, duplicates="drop"
        )
    except ValueError as e:
        # Fallback for edge cases where 3 quantiles cannot be formed (e.g., too few unique revenue values)
        print(f"Warning: Could not create 3 quantiles for publisher segmentation: {e}. Assigning all to 'Indie/Pequeño'.")
        df["categoria_publisher"] = 'Indie/Pequeño' # Fallback for edge cases
    return df

# Aplicacion de la función al dataframe de publishers
# The function `segmentar_publishers_por_cuantil` returns a DataFrame with the new column 'categoria_publisher'.
publisher_stats = segmentar_publishers_por_cuantil(publisher_stats)

# 3. Mapear la clasificación de vuelta al dataframe principal de videojuegos
# Merge using the correctly named column 'categoria_publisher' from publisher_stats
df_steam = df_steam.merge(publisher_stats[['publisher', 'categoria_publisher']], on='publisher', how='left')

# En caso de que algún juego no tuviera publisher registrado, rellenamos como Indie/Pequeño
# Assign the 'categoria_publisher' column to 'publisher_size' in df_steam for consistency with the prompt's desired output name.
df_steam['publisher_size'] = df_steam['categoria_publisher'].fillna('Indie/Pequeño')

# 4. Verificación del resultado de la segmentación
print("Distribución de videojuegos según el tamaño de su Publisher:")
print("-" * 60)
# Print the value counts of the newly created 'publisher_size' column
print(df_steam['publisher_size'].value_counts())
````

### 📊 Interpretación Económica e Implicancias del Negocio

Con la incorporacion de la variable categórica `publisher_size` introduce en el estudio de la **concentración de mercado** donde las asimetrias en el mercado de las distribuidoras. La economia de la plataforma tiene a coexistencia de miles de competidores atómicos junto a gigantes corporativos da lugar a estructuras industriales altamente polarizadas.

Cruzando las variables del volumen del catalogo (`total_games`) y desempeño comercial acumulado (`total_revenue`), donde la segmentación generada por el código aporta la base empírica para evaluar la intensidad competitiva de Steam. Esta estratificación permite aplicar en las secciones posteriores del trabajo práctico herramientas analíticas avanzadas de la economía industrial, tales como el *Coeficiente de Concentración de las g Firmas ($CR_g$)* o el *Índice de Herfindahl-Hirschman (IHH)*.

Analizar el mercado a través de este filtro tricotómico (Grandes, Medianos e Indies) es fundamental para contrastar una de las hipótesis organizacionales más relevantes del proyecto: si la proliferación masiva de títulos en la categoría *Indie/Pequeño* logra disputar de manera efectiva la generación de valor y el excedente del consumidor, o si, por el contrario, el ecosistema se rige bajo una estructura marcadamente oligopólica.

En este último escenario, un porcentaje mínimo de *Grandes Estudios* capturaría el grueso del ingreso total de la plataforma mediante economías de escala y reconocimiento de marca, relegando a la masa de pequeños creadores a competir por los rendimientos marginales en la "larga cola" (*Long Tail*) comercial.

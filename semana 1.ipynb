```python
import numpy as np
import pandas as pd
```

Ahora necesito programar una matriz

```python
produccion = np.array([[100, 200, 150],
                       [250, 150, 300]])
print("1. Matriz original")
print(produccion)
```

Al tener la matriz hay que espesificar que es producto (columna) y que es planta (fila)

```python
df = pd.DataFrame(produccion,
index=['Planta 1', 'Planta 2'],
columns=['Producto 1', 'Producto 2', 'Producto 3'])
```

En el punto 3 y 4 se pide hacer la suma de columnas y filas

```python
total_planta = df.sum(axis=1)
df['Total Planta'] = total_planta
total_producto = df.iloc[:, :3].sum()
print("Total por producto:", total_producto.to_dict())
```

Por ultimo se pide el porsentaje como se pide en el punto 5 y 6 (4)

```python
total_general = total_planta.sum()
df['Participación porcentual'] = (total_planta / total_general * 100).round(2)
print("Tabla completa")
print(df)
```
